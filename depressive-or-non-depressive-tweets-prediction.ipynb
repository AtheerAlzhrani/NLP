```python
import pandas as pd
import numpy as np
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('/kaggle/input/depressivenon-depressive-tweets-data/clean_tweet_Dec19ToDec20.csv')
```

```python
import nltk
nltk.download('stopwords')
```

```python
port_stem=PorterStemmer()
```

```python
def processing(text):
    p_text=re.sub('[^a-zA-Z]',' ',text)
    p_text=p_text.lower()
    p_text=p_text.split()
    p_text=[port_stem.stem(word)for word in p_text if not word in stopwords.words('english')]
    p_text=' '.join(p_text)
    return p_text
```

```python
data.dtypes
```

```python
data['text'] = data['text'].astype(str)
```

```python
data['text']=data['text'].apply(processing)
```

```python
print(data['text'])
```

```python
X=data['text'].values
y=data['sentiment'].values
```

```python
print(y)
```

```python
vectorizer=TfidfVectorizer()
vectorizer.fit(X)
X=vectorizer.transform(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
acc=accuracy_score(pred,y_test)
```

```python
print('Accuracy score  ',acc)
```

```python
X_new=X_test[3]
predication=model.predict(X_new)
print(predication)

if(predication[0]==0):
    print('The text is Depressive ')
else:
    print('The text is Non Depressive')
```
